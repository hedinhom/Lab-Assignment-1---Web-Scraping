```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time

def setup_driver():
    """
    Set up Chrome WebDriver automatically
    """
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.maximize_window()
    return driver


def scrape_reviews(driver, urls):
    """
    Loop through each review page URL,
    extract review data, and return results
    """
    reviews_data = []

    for url in urls:
        driver.get(url)
        time.sleep(4)

        review_blocks = driver.find_elements(By.CLASS_NAME, "ebay-review-section")

        for review in review_blocks:

            try:
                reviewer = review.find_element(By.CLASS_NAME, "review-item-author").text
            except:
                reviewer = "Not Available"

            try:
                date = review.find_element(By.CLASS_NAME, "review-item-date").text
            except:
                date = "Not Available"

            try:
                content = review.find_element(By.CLASS_NAME, "review-item-content").text
            except:
                content = "Not Available"

            reviews_data.append({
                "Reviewer Name": reviewer,
                "Review Date": date,
                "Review Content": content
            })

    return reviews_data


def save_to_csv(data, filename):
    if len(data) == 0:
        print("No reviews extracted.")
        return

    keys = data[0].keys()

    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        writer.writerows(data)


# ====================== MAIN ======================

review_urls = [
    "https://www.ebay.com/urw/Inkbird-ITC306T-Heat-Temperature-Waterproof-Controller-with-Aquarium-Probe-Sensor-No-Cooling/product-reviews/7032612989?pgn=1",
    "https://www.ebay.com/urw/Inkbird-ITC306T-Heat-Temperature-Waterproof-Controller-with-Aquarium-Probe-Sensor-No-Cooling/product-reviews/7032612989?pgn=2",
    "https://www.ebay.com/urw/Inkbird-ITC306T-Heat-Temperature-Waterproof-Controller-with-Aquarium-Probe-Sensor-No-Cooling/product-reviews/7032612989?pgn=3",
    "https://www.ebay.com/urw/Inkbird-ITC306T-Heat-Temperature-Waterproof-Controller-with-Aquarium-Probe-Sensor-No-Cooling/product-reviews/7032612989?pgn=4",
    "https://www.ebay.com/urw/Inkbird-ITC306T-Heat-Temperature-Waterproof-Controller-with-Aquarium-Probe-Sensor-No-Cooling/product-reviews/7032612989?pgn=5"
]

driver = setup_driver()

all_reviews = scrape_reviews(driver, review_urls)

save_to_csv(all_reviews, "Inkbird_ITC306T_Reviews.csv")

driver.quit()

print("Scraping completed. File saved as Inkbird_ITC306T_Reviews.csv")
```
